```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

# Z-Score

```python

data = [11,10,12,13,15,10,11,18,10,9,13,14,15,9,109,10,12,17,14,15,18,20,11,21,11,13,110,16,16,90 ]


def detect_outliers (dataset):
  threshold = 2 # 2 STD 
  mean = np.mean(dataset)
  std = np.std(dataset)
  #print(mean)
  outliers = []
  for item in dataset:
    z_score = (item-mean) / std
    #print(item , z_score)
    if np.abs(z_score) >= threshold:
      outliers.append(item)
  return outliers
```

```python
plt.plot(data)
```

```python
print(detect_outliers (data))
```

# IQR


1. Sort Dataset
2. Calculate Q1 and Q3 
3. Calculate IQR = Q3  - Q1
4. Calculate Lower Fence (Q1 - 1.5(IQR))
5. Calculate Upper Fence (Q3 + 1.5(IQR))



```python
def detect_outliers_IQR(dataset):
  outliers = []
  
  dataSorted = sorted(dataset)
  Q1 , Q3 = np.percentile(dataSorted , [25 , 75])
  IQR = Q3 - Q1
  lower = Q1 - 1.5 * IQR
  upper = Q3 + 1.5 * IQR


  for item in data:
    if item < lower or item > upper:
      outliers.append(item)

  return outliers
```

```python
print(detect_outliers_IQR(data))
```
